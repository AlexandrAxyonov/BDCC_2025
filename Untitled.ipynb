```python
import pandas as pd

# ----- ДАННЫЕ -----
rows = [
    (1, "Video-Audio-Text", "SSL", "CD", 64.26, 63.54, 92.00, 73.39),
    (2, "Video-Text-Audio", "SSL", "CD", 64.18, 63.40, 92.02, 73.37),
    (3, "Audio-Video-Text", "SSL", "CD", 63.98, 63.25, 92.00, 73.03),
    (4, "Audio-Text-Video", "SSL", "CD", 64.18, 63.35, 92.02, 73.42),
    (5, "Text-Video-Audio", "SSL", "CD", 63.69, 63.13, 92.03, 73.27),
    (6, "Text-Audio-Video", "SSL", "CD", 64.19, 63.45, 92.03, 73.15),
]

columns = [
    "Exp ID", "Modality order", "LT", "LD",
    "MOSEI", "FIv2", "mACC", "mCCC"
]

df = pd.DataFrame(rows, columns=columns)

# ----- РАНГИРОВАНИЕ -----
metric_cols = ["MOSEI", "FIv2", "mACC", "mCCC"]

# Приведение к числам (на случай запятых или пробелов)
for c in metric_cols:
    df[c] = df[c].astype(str).str.replace(",", ".", regex=False).str.strip()
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Ранги по метрикам (большее значение — лучше)
ranks = df[metric_cols].rank(method="min", ascending=False)

# Добавляем ранги по каждой метрике в таблицу
for col in ranks.columns:
    df[f"{col}_Rank"] = ranks[col].astype(int)

# Средний ранг по всем метрикам
df["Avg_Rank"] = ranks.mean(axis=1).round(3)

# Сортировка по среднему рангу (чем меньше — тем лучше)
df_sorted = df.sort_values("Avg_Rank", ascending=True).reset_index(drop=True)

# ----- НАСТРОЙКИ ВЫВОДА -----
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.colheader_justify", "center")

print(df_sorted.to_string(index=False))
```
